Let us start with getting know the dataset. Your first task will be to get some basic information by using Pandas features.

```python
#For each task below, look for a Pandas function to do the task.
#Replace None in each task with your code.

# 2.5 points
#Task 1-a: Print the details of the df_olympics data frame (information such as number of rows,columns, name of columns, etc)
print (">>Task 1-a: Details of df_olympics data frame are: \n")
print(df_olympics.info())

# 2.5 points
#Task 1-b: Find the number of rows and columns in the df_olympics data frame.
num_rows = len(df_olympics.index)
num_cols = len(df_olympics.columns)
print ("\n\n>>Task 1-b: Number of rows:%s and number of columns:%s" % (num_rows, num_cols)) 

# 2.5 points
#Task 1-c: Print the descriptive detail (min, max, quartiles etc) for 'Age' column of  the df_olympics
print ("\n\n>>Task 1-c: Descriptive details of year column are\n",df_olympics['Age'].describe())

# 10 points
#Task 1-d: Print the number of years from the first game until the last in our data-set, and the number of  unique values for 'games'.
num_tot_years  = len(df_olympics['Year'].unique())
num_uniq_games = len(df_olympics['Games'].unique())
print ("\n\n >>Task 1-d: In our dataset we have historical data for %s years, and for %s games. " % (num_tot_years, num_uniq_games))
```

# Task 2: Aggregation & Filtering & Rank
In this task, we will perform some very high level aggregation and filtering operations. 
Then, we will apply ranking on the results for some tasks. 
Pandas has a convenient and powerful syntax for aggregation, filtering, and ranking. 
DO NOT write a for loop. Pandas has built-in functions for all tasks. 

```python
# 8 points
#Task 2-a: Find out the total number of female and male athletes that participated on the 2004 Olympics Games
num_female_2004 = len(df_olympics[(df_olympics['Sex']=='F') & (df_olympics['Year']==2004)]['Name'].unique())
num_male_2004   = len(df_olympics[(df_olympics['Sex']=='M') & (df_olympics['Year']==2004)]['Name'].unique())
print (">>Task 2-a: At the Olympics of 2004, there were participating %s female and %s male athletes" 
       % (num_female_2004, num_male_2004))

# 8 points
#Task 2-b: Find out the total number of awarded metals for the year 1896, and the year 2016.
#Utilize the property of cell's missing values
num_medals_1986 = len(df_olympics[(df_olympics['Medal'].notna()) & (df_olympics['Year']==1896)])
num_medals_2016 = len(df_olympics[(df_olympics['Medal'].notna()) & (df_olympics['Year']==2016)])

print ("\n\n>>Task 2-b: The total number of metals awarded in 1896 was %s, while in 2016 was %s" 
       % (num_medals_1986, num_medals_2016))

# 14 points
#Task 2-c: Find out the top 10 athletes with the most gold medals for all years.
top10_gold_athletes=df_olympics[(df_olympics['Medal']=='Gold')]['Name'].value_counts().sort_values( ascending=False).head(10)
print ("\n\n>>Task 2-c: The top 10 athletes for all years are: \n%s" % (top10_gold_athletes))

```

# Task 3: Visualization
In this task, you will perform a number of visualization tasks to get some intuition about the data. Visualization is a key component of exploration. You can choose to use either Matplotlib or Seaborn for plotting. The default figures generated from Matplotlib might look a bit ugly. So you might want to try Seaborn to get better figures. Seaborn has a variety of styles. Feel free to experiment with them and choose the one you like. We have earmarked 10 points for the aesthetics of your visualizations.

```python
sns.set_style('whitegrid')
sns.set(font_scale = 1.3)

# 15 points
# Task 3-a: Draw a histogram for total number of athletes participated in all Summer Olympic Games. 
# Think of a way to nicely visualize the all years of Summer Olympics! 
#########################begin code for Task 3-a
summer_athletes = df_olympics[df_olympics['Season']=='Summer'][['Year', 'Name']].drop_duplicates().copy()
plot = summer_athletes['Year'].hist(color='#539caf',bins=31)
plot.set_title("Total number of athletes participated in all Summer Olympic Games")
plot.set_xlabel("Year of Summer Olympic Games")
plot.set_ylabel("Number of athletes")
plt.show()
print("\n\n")

#########################end code for Task 3-a


# 15 points
# Task 3-b: Draw a "vertical" bar chart that lists the top-10 talest althlets for all years.
# Remember to make the bar chart into a vertical bar chart
#########################begin code for Task 3-b 
plt.figure(figsize=(10,10))
plt.subplot(211)
tall_athletes=df_olympics.drop_duplicates(subset='Name', keep="last").sort_values(by=['Height'], ascending=False).head(10)
ax = sns.barplot(x=tall_athletes['Name'], y=tall_athletes['Height'],data=tall_athletes,palette=sns.cubehelix_palette(len(tall_athletes['Name'])))
ax.set_xlabel(xlabel='Name of Athletes', fontsize=13)
ax.set_ylabel(ylabel='Height of Athletes', fontsize=13)
ax.set_title(label='Top-10 tallest athletes for all years', fontsize=15)
ax.set_xticklabels(labels =tall_athletes['Name'], fontsize=10, rotation=70)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.show()


#########################end code for Task 3-b
```

# Task 4: 
Find out an 'interesting' information from your Olympic History dataset. Create a visualization for it and explain in a few lines your reasoning. 

This task is worth 20 points. Your result will be judged based on the uniqueness and quality of your work (having a meaningful result and an aesthetic visualization). 

```python
#########################begin code for Task 4
#Graph-1
print("\n\n")
print("Graph-1 : Following graph is a factor plot that displays the number of medals (gold,silver & bronze) won by the Team USA in different sports for all the years. With this graph we can analyze the performance and achievement of Team USA in different sports. From the graph we can also infer that Team USA has won most gold medals in swimming.")
usa_medal=df_olympics[df_olympics['Team']=='United States'].groupby(['Sport','Medal']).size().reset_index(name='Medals Count')
f=sns.factorplot("Sport", "Medals Count", "Medal", data=usa_medal, kind="bar", palette="Set1", legend=False,size=5, aspect=3)
f.set_xticklabels(fontsize=12, rotation=85)
f.set_yticklabels(fontsize=12)
plt.title("Graph 1 - Number of Medals won by team USA for all years in different sports")
f.despine(left=True)
plt.legend(loc='upper right')
plt.show()
print("\n\n\n")

#Graph-2
print("Graph-2 : Following graph is a point plot that displays the number of Males and Females who have particpated in Summer Olympics in all Years. With this graph we can compare the number of male to the female particpants every summer Olympics. From the graph we can also infer that number of women who particpated in Summer Olympics is always lower than the number of males.")
sns.set(font_scale = 1.3)
plt.figure(figsize=(10,10))
f, ax= plt.subplots(figsize=(20,10))
sex_cnt= df_olympics[df_olympics['Season']=='Summer'].groupby(['Year','Sex']).size().drop_duplicates().reset_index(name='Number Of Athletes')
ax=sns.pointplot(x="Year", y="Number Of Athletes", hue="Sex", data=sex_cnt, ax= ax,markers=["o", "o"],palette="Set1", scale=0.8)
ax.set_title("Graph 2  - Athletes in Summer Olympics based on Gender",fontsize=20)
ax.set_xlabel(xlabel='Year of Summer Olympics', fontsize=15)
ax.set_ylabel(ylabel='Number of Male & Female Particpants', fontsize=15)
plt.show()
print("\n\n")
#########################end code for Task 4
```
